# Plots

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.array(np.arange(0,0.99,0.01))
y1 = np.sin(2*np.pi*4*t)
y2 = np.cos(2*np.pi*4*t)
```

```python
fig = plt.figure(figsize=(30,20), dpi=20)
ax = fig.add_subplot(111)
ax.plot(t,y1,color='b',linewidth=3)
ax.plot(t,y2,color='r',linewidth=3)
ax.tick_params(axis='both', which='major', labelsize=40)
ax.set_title("sin/cos", fontsize=50)
ax.set_xlabel("t", fontsize=50)
ax.set_ylabel("f(t)", fontsize=50)
```
