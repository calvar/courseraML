## =================== Part 3: Cost and Gradient descent ===================

The cost function for the regression of the m training examples is
$$J(\theta_0,\theta_1)=\frac{1}{2m}\sum_{i=1}^m\left(y^{(i)}-(\theta_0x_0^{(i)}+\theta_1x_1^{(i)})\right)^2=\frac{1}{2m}\sum_{i=1}^m\left(y^{(i)}- \vec{x}^{(i)}\cdot\vec{\theta}\right)^2,$$
where the vector
$$\vec{\theta}=\left(\begin{array}{c}
\theta_0\\
\theta_1\\
\end{array}\right)$$
holds the fitting parameters of the regression. The training examples are stored in a matrix $X$ and a vector $y$:
$$X=\left(\begin{array}{cc}
x_0^{(1)} & x_1^{(1)}\\
x_0^{(2)} & x_1^{(2)}\\
\vdots & \vdots\\
x_0^{(m)} & x_1^{(m)}
\end{array}\right),\ \ 
\vec{y}=\left(\begin{array}{c}
y^{(1)}\\
y^{(2)}\\
\vdots\\
y^{(m)}
\end{array}\right)$$
with $x_0^{(i)}=1$ for all $i$. We refer to the column $k$ of $X$ as $\vec{x}_k$ and to row $i$ of $X$ as $\vec{x}^{(i)}$. We have then
$$J(\vec{\theta})=\frac{1}{2m}\sum_{i=1}^m\left(y^{(i)}-\vec{x}^{(i)}\cdot\vec{\theta}\right)^2=\frac{1}{2m}\left(\vec{y}-X\cdot\vec{\theta}\right)^T\cdot\left(\vec{y}-X\cdot\vec{\theta}\right).$$

```python
# Add column of ones to X
data.insert(0, 'x0', [1 for i in range(m)])
data.head()
```

```python
#Extract matrix X and convert to array
X = data.iloc[:,0:2].to_numpy()
#Extract vector y as a column vector
y = data.iloc[:,2].to_numpy()[np.newaxis].T #create a new axis to convert a 1d array to 2d and transpose
```

The gradient of $J$ with respect to $\theta$ is
$$\frac{\partial J}{\partial\theta_k}=-\frac{1}{m}\sum_{i=1}^m\left(y^{(i)}-\vec{x}^{(i)}\cdot\vec{\theta}\right)x_k^{(i)}=-\frac{1}{m}\vec{x}_k^T\cdot\left(\vec{y}-X\cdot\vec{\theta}\right),$$
where $.^T$ denotes the transpose. Gradient descent consists in displacing the vector $\theta$ in the direction negative to the gradient of the cost function, to look for the minimum. That is
$$\theta_k=\theta_k-\alpha\frac{\partial J}{\partial\theta_k}$$
for all $k$, where $\alpha$ is a convergence parameter.<br /> <br />
At any iteration the cost function $J$ can be used to estimate the error.

```python
#Cost funtion J
def J(X, y, theta):
    v1 = y-np.matmul(X,theta)
    return np.matmul(v1.T,v1)[0][0]/(2*len(y))

#Gradient of J
def grad_J(X, y, theta):
    v1 = y-np.matmul(X,theta)
    grad = []
    for k in range(X.shape[1]):
        tcol = X[:,k] #k column as 1d array (row vector)
        grad.append(np.matmul(tcol,v1))
    return -np.asarray(grad)/len(y)

#gradient descent
def grad_desc(alpha, X, y, epochs, iprint):
    #Initializa parameter vector (fitting parameters)
    theta = np.zeros((2,1))
    for e in range(epochs):
        theta = theta - alpha*grad_J(X,y,theta)
        if e%iprint == 0:
            print(theta, J(X,y,theta))
    return theta
```

```python
#Gradient descent settings
iterations = 1500
alpha = 0.01

THETA = grad_desc(alpha, X, y, iterations, 100)
print("Theta found by gradient descent: ({0:.4f}, {1:.4f})".format(THETA[0][0],THETA[1][0]))
print("Espected theta values (approx.): (-3.6303, 1.1664)")
```

```python
#Plot the linear fit
fig, ax = plt.subplots(figsize=(15, 6))
data.plot.scatter(ax=ax, title="data", x="x1", y="y")
xx = [i for i in np.arange(5,23,1)]
yy = [THETA[0][0]+THETA[1][0]*i for i in xx]
ax.plot(xx,yy,color='r')
```

```python
#Predict values for population sizes 35000 and 70000
p1 = np.array([1, 3.5])
predict1 = np.matmul(p1,THETA)[0]
print("For population = 35000, we predict a profit of ${0:.3f}".format(predict1*10000))

p2 = np.array([1, 7])
predict2 = np.matmul(p2,THETA)[0]
print("For population = 70000, we predict a profit of ${0:.3f}".format(predict2*10000))
```

## ============= Part 4: Visualizing J($\theta_0$, $\theta_1$) =============

```python
#Grid over which we will calculate J
theta0_vals = [i for i in np.arange(-10,10,0.2)]
theta1_vals = [j for j in np.arange(-1,4,0.05)]
J_vals = np.array([[J(X,y,np.array([[i],[j]])) for i in theta0_vals] for j in theta1_vals])
J_vals
```

```python
plt.contour(J_vals,cmap='inferno',extent=[-10,10,-1,4],levels=[i for i in np.arange(0,100,5)])
plt.xlabel(r"$\theta_0$",fontsize=20)
plt.ylabel(r"$\theta_1$",fontsize=20)
plt.colorbar()
```

```python
plt.imshow(J_vals, interpolation='nearest', origin='lower', cmap='inferno', extent=[-10,10,-1,4], aspect='auto', vmin=0, vmax=50)
plt.xlabel(r"$\theta_0$",fontsize=20)
plt.ylabel(r"$\theta_1$",fontsize=20)
plt.colorbar()
```
