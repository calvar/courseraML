## Anomaly detection
We have a system described by a set of features, and a set of data $x_j^{(i)}$, where the subindex represents the feature and the superindex represents the data sample. The algorithm presented assumes that each feature $j$ is Gaussian distributed with mean $\mu_j$ and variance $\sigma_j^2$, and that the featuers are independent. That is, each feature is distributed with probability density
$$p(x_j;\mu_j,\sigma_j)=\frac{1}{\sqrt{2\pi\sigma_j^2}}e^{-\frac{(x_j-\mu_j)^2}{2\sigma_j^2}},$$
while the probability density for a sample $x=\{x_1,x_2,\dots,x_N\}$ is simply
$$p(x)=\prod_{j=1}^Np(x_j;\mu_j,\sigma_j).$$
The idea is that anomalous events $x^{(k)}$ have a small probability of ocurring, so we need to find a value $\epsilon$ so that if
$$p(x)<\epsilon$$
the event is classified as anomalous. Otherwise it is considered normal.

In order to train the model (i.e. estimate the values of the parameters $\mu_j$, $\sigma_j^2$), we can split the data in training, cross validation (CV) and test sets, and then fit the gaussians to the training set. <br>
In order to estimate an appropriate value for $\epsilon$, we use a CV set where the data has been labeled as anomalous ($y=1$) or normal ($y=0$), and adjust $\epsilon$ to have a desired value in some quality parameter. As the data is normaly heavily skewed, as there are much more normal examples than anomalous ones, measuring the accuracy is not very useful, and other ways to asses the quality of the model, that take into account the skewness of the data, are needed. Examples of quality metrics are
- True positive, false positive, true negative, false negative matrix.
- Precision (true positives / true positives + false positives).
- Recall (true positives / true positives + false negatives).
- $F_1$ score. $F_1=2\frac{precision * recall}{precision + recall}$

### Anomaly detection vs. supervised learning
Usually, when there is a classification problem in which the number of data for one class is much less than for the other class, anomaly detection is used. Otherwise, a supervised learning algorithm might be in order. Also, if additionally there might be many different kind of anomalies, with small probability, then anomaly detectionmigh be better. Examples of problems that might be approached with anomaly detection:
- Fraud detection.
- Manufacturing errors.
- Monitoring computers in data centers.

Examples of problems to be approached with supervised learning:
- Email spam classification.
- Weather prediction.
- Cancer classification.

### Non-gaussianity of data
Sometimes the data is distributed in a clearly non-gaussian way (for ex. with a long tail like an exponential). In these cases one can try to transform the data, by using for example a logarithm, or elevating to some non integer power, until the data looks more or less symetric and closer to a gaussian.
### Error analysis
On occasion one finds that data, which on knows is anomalous, falls well within the accepted probability range, that is, the algoritm is not able to detect the anolmaly. In these cases one can use domain knowledge to propose additional features that might be new or combinations of existing features, to expand the phase space in hoopes that the data can be now correctly calssified as anomalous.

### Multivariate distribution
Sometimes we cannot make the assumption of independence between the features. For example, in the surveillance of a cluster of computers the CPU load might be correlated with the memory usage, such that high CPU load usually means high memory usage. In these cases the projection of the data in the individual feature axes might not be able to distinguish anomalies, and we need to consider the multivariate distribution. 
<img src="multidist.png" width=400/>
Sometimes creating new features that combine the correlated features in a non-linear way can do the trick, but it is not always easy to find these combinations.
The multivariate gaussian distribution
$$p(x_1,\dots,x_k;\mu,\Sigma)=\frac{\exp\left(-\frac{1}{2}(x-\mu)\Sigma^{-1}(x-\mu)^T\right)}{\sqrt{(2\pi)^k|\Sigma|}},$$
depends on parameters $\mu=(\mu_1,\dots,\mu_k)$, with
$$\mu_j=\frac{1}{m}\sum_{i=1}^{m}x_j^{(i)},$$
and
$$\Sigma=\text{Cov}[X_1,\dots,X_k]=\frac{1}{m}\sum_{i=1}^{m}(x^{(i)}-\mu)(x^{(i)}-\mu)^T,$$
which is the covariance matrix. $|\Sigma|$ denotes the determinat of $\Sigma$, while $\Sigma^{-1}$ denotes its inverse. The multivariate distribution captures correctly all correlations, but can be significantly more expensive in computational terms. It is important also to know that if the number of samples is smaller than the number of features, then $\Sigma$ will not be inverible.

## ================== Part 2: Estimate the dataset statistics ===================

```python
#Estimate the mean and variance of independent features, assuming Gaussian distribution
def estimateGaussian(X):
    mean = np.array([X[:,i].mean() for i in range(X.shape[1])])
    var = np.array([[X[:,i].var() for i in range(X.shape[1])]])
    return mean, var

#Estimate the pdf of a set of data, assuming Gaussian distribution. If sigma2 is a matrix
# it assumes it is the covariance matrix. If it is a vector it assumes is the diagonal of
# the covariance matrix for independent random variables.
def multivariateGaussian(X, mu, sigma2):
    k = len(mu)
    
    s2 = np.copy(sigma2)
    if sigma2.shape[0] == 1:
        s2 = np.diag(sigma2[0])
    
    num = (2*np.pi)**(k/2) * np.linalg.det(s2)**(0.5)
    Xm = X-mu
    sXm = np.matmul(np.linalg.inv(s2),Xm.T)
    XmsXm = np.array([np.matmul(Xm[i,:], sXm[:,i]) for i in range(Xm.shape[0])])
    p = np.exp(-0.5 * XmsXm) / num
    
    return p
    
```

## ================== Part 3: Find Outliers ===================

```python
#Look for a good epsilon threshold using the validation set
pval = multivariateGaussian(Xval, mu, sigma2)
```

```python
#Look for the epsilon that maximises the F1 score
def selectThreshold(yval, pval):
    yp_df = pd.DataFrame(yval, columns=['y'])
    yp_df['p'] = pval 
    
    bestF1 = 0
    bestEpsilon = 0
    evals = np.linspace(np.min(pval), np.max(pval), 1000)
    for epsilon in evals:
        tp = yp_df[(yp_df.p<epsilon) & (yp_df.y==1)].count() #true positive
        fp = yp_df[(yp_df.p<epsilon) & (yp_df.y==0)].count() #false positive
        fn = yp_df[(yp_df.p>=epsilon) & (yp_df.y==1)].count() #false negative
        
        precision = tp[0] / (tp[0] + fp[0])
        recall = tp[0] / (tp[0] + fn[0])
        F1 = 2 * precision * recall / (precision + recall)
        if F1 > bestF1:
            bestF1 = F1
            bestEpsilon = epsilon
    
    return bestEpsilon, bestF1
```

```python
epsilon, F1 = selectThreshold(yval, pval)
```

```python
print("Best epsilon found using cross-validation: {0}".format(epsilon))
print("Best F1 on Cross Validation Set: {0}".format(F1))
print("(you should see a value epsilon of about 8.99e-05)")
print("(you should see a Best F1 value of  0.875000)")
```

```python
#Find the outliers
X_df['p'] = p
X_df['out'] = X_df['p'].apply(lambda x: x < epsilon)
X_df.head()
```

```python
fig2, ax = plt.subplots(figsize=(8,6))
sns.scatterplot(data=X_df, ax=ax, x='Latency (ms)', y='Throughput (mb/s)', hue='out')
ax.contour(fvals, cmap='inferno', extent=[4,24,4,25], levels=10.0**np.array(range(-20,0,3)))
```

```python
data = loadmat("ex8data2.mat")
Xb = data['X']
Xbval = data['Xval']
ybval = data['yval']
Xb.shape, Xbval.shape, ybval.shape
```

```python
#Estimate mean and variance
mub, sigma2b = estimateGaussian(Xb)
mub, sigma2b
```

```python
#Compute probabilities for the training set
pb = multivariateGaussian(Xb, mub, sigma2b)
pb.shape
```

```python
#Compute probabilities for the cross-validation set
pvalb = multivariateGaussian(Xbval, mub, sigma2b)
pvalb.shape
```

```python
#Find epsilon that maximizes the F1 score 
epsilonb, F1b = selectThreshold(ybval, pvalb)
```

```python
print("Best epsilon found using cross-validation: {0}".format(epsilonb))
print("Best F1 on Cross Validation Set: {0}".format(F1b))
print("(you should see a value epsilon of about 1.38e-18)")
print("(you should see a Best F1 value of 0.615385)")
```

```python
#Find the outliers
Xb_df = pd.DataFrame(Xb)
Xb_df['p'] = pb
Xb_df['out'] = Xb_df['p'].apply(lambda x: x < epsilonb)
Xb_df.head()
```
